```python
import requests
from bs4 import BeautifulSoup

url = "https://matapropertybali.com/property/villa-for-sale-in-balangan-walking-distance-to-the-beach/"

response = requests.get(url, timeout=15)
response.raise_for_status()

soup = BeautifulSoup(response.text, "html.parser")

# Title
title_tag = soup.select_one(".page-title h1")
title = title_tag.get_text(strip=True) if title_tag else ""

# Price (only if title exists)
price = ""
if title:
    price_tag = soup.select_one("li.item-price")
    price = price_tag.get_text(strip=True) if price_tag else ""

print("Title:", title)
print("Price:", price)
```
